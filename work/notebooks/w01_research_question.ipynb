## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Decision:
The goal is to decide which content pages should be prioritized for refresh based on their search performance and content metrics.

Action:
The content or SEO team will review and update the highest-priority pages first instead of spending time reviewing every page manually.

Cost of a Wrong Call:
If a page that does not need attention is prioritized, time and resources are wasted. If an important page is missed, it may continue losing search visibility, traffic, and user engagement.

Why this decision matters:
The dataset contains performance indicators such as impressions, clicks, CTR, engagement, and content age, which can help identify pages that are more likely to benefit from a content refresh.
